```python
import math
import random
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

<h2>Monty Hall Problem</h2>

<h3>Functions</h3>

```python
# 
def randomizeDoor():
    return random.randint(1,3)

# 
def hostChooseDoor(correct_door, player_door):
    door_list = [1, 2, 3]
    door_list.remove(correct_door)
    if correct_door != player_door:
        door_list.remove(player_door)
    doorNum = random.choice(door_list)
    return doorNum

# 
def finalPlayerChoice(og_choice, open_door, player_num):
    if player_num == 1: return og_choice
    door_list = [1, 2, 3]
    door_list.remove(open_door)
    if player_num == 2: door_list.remove(og_choice)
    return random.choice(door_list)

# 
def isVictory(correct_door, player_door):
    return correct_door == player_door

# 
def montyHallGame(player_num):
    correct_door = randomizeDoor()
    player_door = randomizeDoor()
    return isVictory(correct_door, finalPlayerChoice(player_door, hostChooseDoor(correct_door, player_door), player_num))

# 
def multipleMontyHallGamesForPlayer(amount, player_num):
    return [montyHallGame(player_num) for i in range(amount)]

# 
def calcWinRatio(results):
    return [sum(results[:i])/i for i, _ in enumerate(results, start=1)]

# 
def plotGraphForMontyHallOfPlayer(amount, player_num):
    results = multipleMontyHallGamesForPlayer(amount, player_num)
    ratio = calcWinRatio(results)
    ratio = [0.0] + ratio
    plt.plot(range(amount+1), ratio, color='darkorange')
    plt.xlim([0.0, amount])
    plt.ylim([0.0, 1.05])
    plt.xlabel(f'Number of Games Played')
    plt.ylabel(f'Win Ratio')
    plt.title(f'Win Ratio of Player {player_num}')
    plt.show()
    return

# 
def plotGraphForMontyHall(amount):
    plotGraphForMontyHallOfPlayer(amount, 1)
    plotGraphForMontyHallOfPlayer(amount, 2)
    plotGraphForMontyHallOfPlayer(amount, 3)
    return
```

```python
amount = 1000
```

```python
plotGraphForMontyHallOfPlayer(amount, 1)
```

```python
plotGraphForMontyHallOfPlayer(amount, 2)
```

```python
plotGraphForMontyHallOfPlayer(amount, 3)
```

We can see that the player who had the higher win ratio is player 2 with about 0.66666, whereas player one had about 0.33333 and player 3 had about 0.5.
